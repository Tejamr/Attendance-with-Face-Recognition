```python
import numpy as np
import cv2
import face_recognition
```

```python
img1 = face_recognition.load_image_file('Images/Modi.jpeg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)


img2 = face_recognition.load_image_file('Images/chiru.jpeg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
faceloc = face_recognition.face_locations(img1)[0]
encoding1 = face_recognition.face_encodings(img1)[0]
cv2.rectangle(img1,(faceloc[3],faceloc[0],faceloc[1],faceloc[2]), (255,0,255),2)

faceloc = face_recognition.face_locations(img2)[0]
encoding2 = face_recognition.face_encodings(img2)[0]
cv2.rectangle(img2,(faceloc[3],faceloc[0],faceloc[1],faceloc[2]), (255,0,255),2)
```

```python
results = face_recognition.compare_faces([encoding1],encoding2)


facedistance = face_recognition.face_distance([encoding1],encoding2)

print(results,facedistance)


cv2.putText(img2,f'{results} {round(facedistance[0],2)}' , (50,50) ,cv2.FONT_HERSHEY_COMPLEX,1,(255,0,0),2)
```

```python
cv2.imshow('Modi',img1)

cv2.imshow('Chiru',img2)

cv2.waitKey(0)
```
